These are visualizations of attention weights. When a model translates a sentence, at each generation step it "pays attention" to some words more than to others. The first image above illustrates what we would intuitively expect a model to do: it knows to look more closely at certain words, the translations of which it generates at the moment, while also directing some dispersed attention to other words in the sentence for context. The second image illustrates a bad translation: the model doesn't know where to look, and the attention ends up being spread more or less evenly between all words.

In order to generate attention visuals for sentences, you can either use the default functionality of Sockeye (add `--output-type align_plot` to your translation command, see documentation of `sockeye-translate` command for details), or https://github.com/M4t1ss/SoftAlignments (might need to do `sudo apt get php` and `export BROWSER=google-chrome` if you are using Chrome) that runs a web server locally.

Note that the alignment plot will show subword units and not full words, as this is the representation used by Sockeye during translation.
